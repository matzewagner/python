```python
%pylab inline
import scikits.audiolab as skal
```

```python
dur = 60.0
sr = 44100
t = arange(0, dur, 1.0/sr)
```

```python
carFreq = 2.9
modFreq = 20000
modAmount = 0.9
AMModFreq = 1000
AMMod = sin(AMModFreq*2*pi*t)*t*0.5
sqrMod=0
for x in range(1,22,2):
    sqrMod += sin(2*pi*(modFreq*x)*t)/x

mod = (sin(modFreq*2*pi*t)/dur)*t*(modAmount)
sqrCar=0
for x in range(1,22,2):
    sqrCar += sin(2*pi*(carFreq*x)*mod*t)/x

car = sin(carFreq*mod*2*pi*t)*0.5
```

```python
start = 0
stop = 1000
plot(car[start:stop])
```

```python
width = 512
height = 256
start = int(0.0*sr)
totalLength = width*height*3
sigImg = car[start:(totalLength+start)].reshape(height,width,3)
```
