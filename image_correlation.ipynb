```python
%pylab inline 
from __future__ import print_function
from __future__ import division
from scipy.signal import correlate2d
rcParams['figure.figsize'] = (10, 5)
```

### load black polka dots over white background, then sum each pixel to get the brightness

```python
gs = GridSpec(1,2)
subplot(gs[0,0])
img1 = imread('polka1.png')
imshow(img1)
subplot(gs[0,1])
img1sum = sum(img1.astype(float), axis=2)/len(img1[0,0,:])
imshow(img1sum)
colorbar()
```

### load white polka dots over black background, then sum each pixel to get the brightness

```python
gs = GridSpec(1,2)
subplot(gs[0,0])
img2 = imread('polka2.png')
imshow(img2)
subplot(gs[0,1])
img2sum = sum(img2, axis=2)/len(img2[0,0,:])
imshow(img2sum)
colorbar()
```

### load single black polka dot over white background, then sum each pixel to get the brightness

```python
gs = GridSpec(1,2)
subplot(gs[0,0])
img3 = imread('blackdot.png')
imshow(img3)
subplot(gs[0,1])
img3sum = sum(img3, axis=2)/len(img3[0,0,:])
imshow(img3sum)
colorbar()
```

### cross correlate the black polka dots with the single dot

```python
cc2 = correlate2d(img1sum, img3sum, mode='same')
imshow(cc2)
colorbar()
```

```python
imshow(where(cc2 > 70, 1, 0), cmap=cm.gray)
colorbar()
```

###### As expected, the highest values appear where the dots in the larger image are located. There seems to be some tapering towards the outer edges of the larger image, as well as some bleed from the white background of the single dot (better visible in the heatmap colorscheme).

### Now we correlate the white polka dots over the black background with the single black polka dot. 

```python
cc3 = correlate2d(img2sum, img3sum, mode='same')
imshow(cc3)
colorbar()
```

```python
imshow(where(cc3 > 20, 1, 0), cmap=cm.gray)
colorbar()
```

###### As expected, we can draw the negative image from the original, albeit the white background in the single dot, causes some blurring.

### Next, we correlate the original black dots on white with a 5x5 pixel black image.

```python
gs = GridSpec(1,2)
subplot(gs[0,0])
img4 = imread('black.png')
imshow(img4)
subplot(gs[0,1])
img4sum = sum(img4, axis=2)/len(img4[0,0,:])
imshow(img4sum)
colorbar()
```

```python
cc4 = correlate2d(img1sum, img4sum, mode='same')
imshow(cc4, cmap=cm.gray)
colorbar()
```

###### This allows the original image to be represented fairly accurately. 

### Lets correlate the original black dots over white background, with its negative. 

```python
cc5 = correlate2d(img1sum, img2sum)
imshow(cc5)
colorbar()
```

```python
imshow(cc5, cmap=cm.gray)
colorbar()
```

###### This creates a nice interference pattern, which tapers off towards the outer borders.

### If we cross correlate this pattern with the original image, it will blurr out the definition further, so it almost looks like a sphere (or something under an electron microscope?)

```python
cc6 = correlate2d(cc5, img1sum)
imshow(cc6, cmap=cm.gray)
colorbar()
```

```python
gs = GridSpec(1,2)
subplot(gs[0,0])
board1 = imread('board1.png')
imshow(board1)
subplot(gs[0,1])
board1sum = sum(board1, axis=2)/len(board1[0,0,:])
imshow(board1sum)
colorbar()
```

###### correlating with a checkerboard has a smoothing effect to the sphere

```python
cc7 = correlate2d(cc5, board1sum)
imshow(cc7, cmap=cm.gray)
colorbar()
```

### correlating the checkerboard with itself also produces a nice interference pattern (this ask for more!)

```python
cc8 = correlate2d(board1sum, board1sum)
imshow(cc8, cmap=cm.gray)
colorbar()
```
