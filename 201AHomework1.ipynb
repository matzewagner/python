SQL query from data.stackexchange.com : (http://data.stackexchange.com/stackoverflow/revision/266722/348320/getting-a-little-emotional)

SELECT cast(floor(cast(CreationDate as float)) as datetime),
SUM(CASE WHEN Title like '%stupid%' THEN 1 ELSE 0 END) AS stupid,
SUM(CASE WHEN Title like '%dumb%' THEN 1 ELSE 0 END) AS dumb,
SUM(CASE WHEN Title like '%horrible%' THEN 1 ELSE 0 END) AS horrible
FROM posts
WHERE (Title LIKE '%wrong%' OR Title LIKE '%stupid%' OR Title LIKE '%dumb%'
OR Title LIKE '%horrible%' OR Title LIKE '%slow%' 
AND cast(floor(cast(CreationDate as float)) as datetime) < '2013-12-31')
GROUP BY cast(floor(cast(CreationDate as float)) as datetime)
ORDER BY cast(floor(cast(CreationDate as float)) as datetime)
