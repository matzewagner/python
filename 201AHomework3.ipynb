```python
%pylab inline
from __future__ import division
from __future__ import print_function
from scipy.io import wavfile
```

```python
img = imread('gradientblur.png')
figure(figsize=(10,10))
imshow(img)
img.shape
```

```python
sig = img[:,:,:]
```

```python
red_channel=[]
for each_red_value in img[:,:,0] :
    red_channel = concatenate([red_channel, each_red_value])

green_channel=[]
for each_green_value in img[:,:,1] :
    green_channel = concatenate([green_channel, each_green_value])

blue_channel=[]
for each_blue_value in img[:,:,2] :
    blue_channel = concatenate([blue_channel, each_blue_value])
    
    
brightness_channel = (red_channel + green_channel + blue_channel) / 3

plot(brightness_channel)
len(brightness_channel)
```

```python
win_len = len(sig[1])
win_list = arange(0, len(brightness_channel), win_len)
newIFFT = []

mag_spectrum = []

for row in win_list:
    win = brightness_channel[row: row + win_len]
    X = fft.irfft(win)
    #X = X[0:320]
    newIFFT = concatenate([newIFFT, X])

figure(figsize=(20,5))
plot(newIFFT)
```

```python
print(max(newIFFT), min(newIFFT))
```

```python
outSig = ((newIFFT - min(newIFFT))*2) / (max(newIFFT) - min(newIFFT)) - 1
figure(figsize=(20,5))
plot(outSig)
print(max(outSig), min(outSig))
```

```python
wavfile.write('IFFT.wav', 44100, outSig)
```
