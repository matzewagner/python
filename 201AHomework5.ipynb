## Creating a hybrid texture from 2 patterns using convolution and morphological gradient

```python
%pylab inline
from __future__ import division
from __future__ import print_function
from scipy.ndimage.filters import convolve as convol2
import scipy.ndimage as ndimage
import scipy.ndimage.morphology as morph
```

### load both images, and sum the color values to grayscale

```python
img1 = imread('pattern2.png')
imshow(img1,  cmap=cm.gray, interpolation='nearest')
img1.shape
```

```python
img1Sum = sum(img1, axis=2)/len(img1[0,0,:])
imshow(img1Sum, cmap=cm.gray, interpolation='nearest')
```

```python
img2 = imread('stripes.png')
imshow(img2,  cmap=cm.gray, interpolation='nearest')
img2.shape
```

```python
img2Sum = (sum(img2.astype(float),axis=2)/len(img2[0,0,:]))
imshow(img2Sum, cmap=cm.gray, interpolation='nearest')
```

### Convolve both images

```python
con = convol2(img1Sum, img2Sum)
figure(figsize=(10,10))
imshow(con, cmap=cm.gray, interpolation='nearest')
```

### use morphological gradient to bring out contrast in convolved image

```python
mg = morph.morphological_gradient(con, (3,3))
figure(figsize=(10,10))
imshow(mg, cmap=cm.gray)
```
