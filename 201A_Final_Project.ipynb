## Sinusoidal modeling

The morphing of audio signals is different in both procedure and result from a simple amplitude crossfade. Whereas the latter also describes a transition from one sound to another, the human ear can generally clearly detect two different sounds - one appearing, while the other disappears. A sinusoidal model analyzes the spectral content of the signals, and interpolates between them. Ideally, this leads to the effect of transitioning between different signals as one smooth sound.  

In more detail, in order to attain a sinusoidal model of a signal, we first perform a spectral analysis via the Fourier transform. As a morph between signals unfolds over time, we need to make use of the discrete Fourier transform (preferably FFT). From the magnitude and phase spectrum we can infer peaks, representing sinusoids that are present at that frequency. The sinusoidal model treats the spectrum as a **sum of time-varying sinusoids** (meaning sinusoids, where frequency and amplitude can vary over time).

Peak **tracks** are created, which interpolate between the changing sinusoids in each window. The size of the analysis window also determines the frequency resolution, and therefor has significant impact on the result. The algorithm for defining a track between sinusoids needs to take into account how stable frequency, amplitude and phase are over successive frames. These tracks are then resynthesized as a sum of sine oscillators, to represent the original signal. The actual morph is a time based interpolation between the tracks of multiple signals.

## Loris

The Loris library generally follows this procedure, but expands on it with the *Reassigned Bandwidth-Enhanced Additive Sound Model*. I short, this embodies that each track has a certain bandwidth that exceeds that of a pure sinewave, and includes some amount of noise in addition. Loris offers an extensive yet simple to use toolkit for analyzing signals and morphing between them. 

### Analysis

In order to obtain a satisfactory model of a sound, several things have to be considered during the analysis process. First, we should give estimates of a fundamental frequency and the duration of the sound to make the computations more effective. We can set for frequency resolution, as well as the frequency drift parameter. The latter predetermines how much sinusoids can deviate over time, and is usually set to half of the frequency resolution. In addition to setting a window size, we can also select the hop-time, which constitutes the overlap of analysis windows. Finally, we can use functions to specify frequency floor and amplitude floor, which can help lessening unwanted noise.

### Morph

In addition to the analysis and morph functions, Loris includes several audio editing options, which can be performed prior to the actual morph.  These include the ability to align the onsets of the signals, as well as a time stretching algorithm, which can be performed at an arbitrary number of breakpoints within the audio sample. There is also a pitch shifting function that provides the possibility of bringing signals with different fundamentals closer together. The morph function in Loris simply asks for time values that determine when and over what duration the morph takes place. However, we can define 3 different envelopes, to morph frequency, amplitude and noisiness independently.
